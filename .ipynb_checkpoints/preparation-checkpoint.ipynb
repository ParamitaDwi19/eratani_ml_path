```python
dateparse = lambda x: pd.datetime.strptime(x, '%d-%m-%Y')

df1 = pd.read_csv('Climate_Sby_2013-2021.csv', header=0, parse_dates=[0], date_parser=dateparse)

df1.head(15)
```

```python
df1.tail()
```

```python
df1['Tanggal'].dtype
```

Create separate datasets for each features: Tavg (average temperature), Tn (minimum temperature), Tx (maximum temperature) and RH_avg (average humidity)

```python
df_Tavg = df1[['Tanggal', 'Tavg']]
df_Tn = df1[['Tanggal', 'Tn']]
df_Tx = df1[['Tanggal', 'Tx']]
df_RHavg = df1[['Tanggal', 'RH_avg']]

df_Tavg.head()
```

```python
df_Tn.head()
```

```python
df_Tx.head()
```

```python
df_RHavg.head()
```

## Missing Handle Value

```python
df_Tavg = df_Tavg.fillna(method ='pad')
df_Tn = df_Tavg.fillna(method ='pad')
df_Tx = df_Tavg.fillna(method ='pad')
df_RHavg = df_Tavg.fillna(method ='pad')
```

```python
df_Tavg.to_csv('Tavg_dataset.csv', index=False)
df_Tn.to_csv('Tn_dataset.csv', index=False)
df_Tx.to_csv('Tx_dataset.csv', index=False)
df_RHavg.to_csv('RHavg_dataset.csv', index=False)
```
