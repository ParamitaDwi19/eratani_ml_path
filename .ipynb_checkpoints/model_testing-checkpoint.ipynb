### Handle Missing Values

```python
df3 = df2[['Tanggal', 'Tavg']]
df3.head()
```

```python
df3.info()
```

```python
df3 = df3.fillna(method ='pad')
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
df3.index = df3['Tanggal']
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result_a = seasonal_decompose(df3['Tavg'], model='multiplicative')
result_a.plot()
```

```python
model = SARIMAX(df3['Tavg'], order=(5,1,3), seasonal_order=(1,1,1,12))
```

```python
model_fit = model.fit()
```

```python
residuals = model_fit.resid
```

```python
residuals.plot()
```

```python
output = model_fit.forecast()
```

```python
output
```

```python
model_fit.forecast(28)
```
